Fill the rows in Type_2 column that are currently null with None

```
def fill_type_2(cols):
  type_2 = cols[0]
  if pd.isnull(type_2):
    return "None"
  else:
    return type_2
```

```
df['Type_2'] = df[['Type_2']].apply(fill_type_2, axis = 1)
```
